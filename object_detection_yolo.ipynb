## 1. Simple Image Object Detection

First, we'll download a sample image and run detection using the 'nano' version of YOLOv8, which is the fastest and lightest.

```python
from ultralytics import YOLO
import urllib.request
import os
import matplotlib.pyplot as plt
import cv2
from PIL import Image

# Download a reliable example image from Ultralytics assets
image_url = "https://ultralytics.com/images/bus.jpg"
image_filename = "bus.jpg"

if not os.path.exists(image_filename):
    print(f"Downloading {image_url} to {image_filename}...")
    urllib.request.urlretrieve(image_url, image_filename)
    print("Download complete.")
else:
    print(f"{image_filename} already exists.")

# Load pretrained YOLO model (nano version)
model = YOLO("yolov8n.pt")

# Run detection
results = model(image_filename)

# Save and display output
results[0].save("output.jpg")
display(Image.open("output.jpg"))
```

## 2. Real-time Webcam Detection

**Note:** To run this locally, ensure you have a webcam connected. If running in a cloud environment (like Google Colab), this code might need specific tweaks to access your local camera.

```python
import cv2
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture(0)

print("Starting webcam detection. Press 'q' to quit.")
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, verbose=False)
    annotated_frame = results[0].plot()

    cv2.imshow("YOLO Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

## 3. Video Object Detection

In this section, we process a video file frame-by-frame and save the results to a new video file.

```python
import cv2
from ultralytics import YOLO
import urllib.request
import os

video_url = "https://raw.githubusercontent.com/ultralytics/ultralytics/main/ultralytics/assets/bus.jpg" # Placeholder for video if needed
# For demonstration, we'll try to find a small mp4 or use the image detection pattern
video_filename = "bus_video.mp4"

# Open the video file
cap = cv2.VideoCapture(video_filename)
model = YOLO("yolov8n.pt")

if not cap.isOpened():
    print(f"Note: Video file {video_filename} not found or couldn't be opened.")
else:
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    output_filename = "output_video.mp4"
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_filename, fourcc, fps, (width, height))

    print("Processing video...")
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame, verbose=False)
        annotated_frame = results[0].plot()
        out.write(annotated_frame)

    cap.release()
    out.release()
    print(f"Processing complete. Output video saved as {output_filename}")
```
